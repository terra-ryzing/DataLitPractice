```
import tweepy
import pandas as pd
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```
#My Twitter API Authentication Variables
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)

tweets = api.search('Machine Learning', count=500)


data = pd.DataFrame(data=[tweet.text for tweet in tweets], columns=['Tweets'])

display(data.head(10))


print(tweets[0].created_at)
```

```
import nltk
nltk.downloader.download('vader_lexicon')
```

```
sid = SentimentIntensityAnalyzer()


listy = []

for index, row in data.iterrows():
  ss = sid.polarity_scores(row["Tweets"])
  listy.append(ss)
  
se = pd.Series(listy)
data['polarity'] = se.values

display(data.head(100))
```

```
import matplotlib.pyplot as plt
import numpy as np
```

```
neg, neu, pos = 0,0,0
```

```
for s in se:
  neg += s['neg']
  neu += s['neu']
  pos += s['pos']
```

```
labels = ['Negative', 'Neutral', 'Positive']
index = np.arange(len(labels))
plt.bar(index,[neg,neu,pos],color=['red','black','green'])
plt.ylabel('Count of Sentiments', fontsize=15)
plt.xlabel('Polarity', fontsize=15)
plt.xticks(index, labels, fontsize=15, rotation=30)
plt.title('Sentiment analysis of tweets for the keyword Machine Learning')
```
